## Q1. Given a valid (IPv4) IP address, return a defanged version of that IP address.

A defanged IP address replaces every period "." with "[.]".

Example 1:

Input: address = "1.1.1.1"
Output: "1[.]1[.]1[.]1"
Example 2:

Input: address = "255.100.50.0"
Output: "255[.]100[.]50[.]0"

```python
class Solution:
    def defangIPaddr(self, address):
        a=''
        for i in address:
            if i!='.':
                a+=i
            else:
                a+='[.]'
        return a
a=Solution()
print(a.defangIPaddr("1.1.1.1"))
```

## Q2 Given an array nums of integers, return how many of them contain an even number of digits.

Example 1:

Input: nums = [12,345,2,6,7896]

Output: 2

Explanation: 

12 contains 2 digits (even number of digits). 
345 contains 3 digits (odd number of digits). 
2 contains 1 digit (odd number of digits). 
6 contains 1 digit (odd number of digits). 
7896 contains 4 digits (even number of digits). 
Therefore only 12 and 7896 contain an even number of digits.

```python
class Solution:
    def findNumbers(self, nums): #List[int]) -> int:
        result=[]
        count=0
        for i in range(len(nums)):
            digits=0
            n=nums[i]
            while n>0:
                n=n//10
                digits+=1
            result.append(digits)
                        
        for i in result:
            if i%2==0:
                count+=1
        return count
        
a=Solution()
print(a.findNumbers(nums= [12,345,2,6,7896] ))
```

## Q3 Given an array of integers nums, return the number of good pairs.

A pair (i, j) is called good if nums[i] == nums[j] and i < j.

 

Example 1:

Input: nums = [1,2,3,1,1,3]
Output: 4
Explanation: There are 4 good pairs (0,3), (0,4), (3,4), (2,5) 0-indexed.

```python
class Solution:
    def numIdenticalPairs(self, nums): #List[int]) -> int:
        good=0
        dic={}
        for key in nums:
            if key not in dic:
                dic[key]=1
            else:
                dic[key]+=1
        for i in dic:
            a=dic[i]
            good+=((a*(a-1))//2)
        return good
a=Solution()
print(a. numIdenticalPairs(nums=[1,2,3,1,1,3]))
```

## Q4 Given the array nums, for each nums[i] find out how many numbers in the array are smaller than it. That is, for each nums[i] you have to count the number of valid j's such that j != i and nums[j] < nums[i].

Return the answer in an array.

Example 1:

Input: nums = [8,1,2,2,3]
Output: [4,0,1,1,3]
Explanation: 
For nums[0]=8 there exist four smaller numbers than it (1, 2, 2 and 3). 
For nums[1]=1 does not exist any smaller number than it.
For nums[2]=2 there exist one smaller number than it (1). 
For nums[3]=2 there exist one smaller number than it (1). 
For nums[4]=3 there exist three smaller numbers than it (1, 2 and 2).

```python
class Solution:
    def smallerNumbersThanCurrent(self, nums): #List[int]) -> List[int]:
        output=[]
        n=len(nums)
        for i in range(len(nums)):
            count=0
            for j in range(n):
                if nums[i]>nums[j]:
                    count+=1
            output.append(count)
        return output
a=Solution() 
print(a.smallerNumbersThanCurrent(nums= [8,1,2,2,3]))
```

## Q5 Given an integer number n, return the difference between the product of its digits and the sum of its digits.
 
Example 1:

Input: n = 234
Output: 15 
Explanation: 
Product of digits = 2 * 3 * 4 = 24 
Sum of digits = 2 + 3 + 4 = 9 
Result = 24 - 9 = 15

```python
class Solution:
    def subtractProductAndSum(self, n): #int) -> int:
        prod=1
        add=0
        while n>0:
            a=n%10
            n=n//10
            prod*=a
            add+=a
        return prod-add

a=Solution() 
print(a.subtractProductAndSum(234))
```

## Q6 You are given an integer n and an integer start.

Define an array nums where nums[i] = start + 2 * i (0-indexed) and n == nums.length.

Return the bitwise XOR of all elements of nums.

 

Example 1:

Input: n = 5, start = 0
Output: 8
Explanation: Array nums is equal to [0, 2, 4, 6, 8] where (0 ^ 2 ^ 4 ^ 6 ^ 8) = 8.
Where "^" corresponds to bitwise XOR operator.

```python
class Solution:
    def xorOperation(self, n,start):
        nums=[]
        result=0
        for i in range(n):
            x=start+2*i
            nums.append(x)
        for i in nums:
            result=result^i
        return result
a=Solution()
print(a.xorOperation(n=5,start=0))
```
